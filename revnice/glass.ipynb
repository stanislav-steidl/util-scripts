```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# --- Rozměry (cm) ---
# rozměry hlavní L-plochy
bottom_len = 80  # délka spodního ramene
bottom_depth = 50 # tloušťka spodního ramene
left_height = 100 # výška levého ramene
left_width = 30   # tloušťka levého ramene

for chamfer in  [10,15,20]:  # délka zkosení rohů

    # krb
    krb_x = 30    # odsazení od levé hrany
    krb_y = 50    # odsazení od horní hrany (šedé zdi)
    krb_w = 50
    krb_h = 35

    # --- Kreslení ---
    fig, ax = plt.subplots(figsize=(8,6))

    # L-profil - kreslíme modře
    # levé svislé rameno
    ax.add_patch(patches.Rectangle((0, 0), left_width, left_height, 
                                linewidth=2, edgecolor='none', facecolor='blue', alpha=0.1))
    # horní horizontální úsek
    #ax.add_patch(patches.Rectangle((left_width, left_height - top_len), 
    #                               top_len, left_width, 
    #                               linewidth=2, edgecolor='blue', facecolor='none'))
    # spodní horizontální úsek
    ax.add_patch(patches.Rectangle((left_width, 0), bottom_len, bottom_depth, 
                                linewidth=2, edgecolor='none', facecolor='blue', alpha=0.1))
    # pravé svislé rameno
    #ax.add_patch(patches.Rectangle((left_width + bottom_len - right_width, 0),
    #                               right_width, right_height,
    #                               linewidth=2, edgecolor='blue', facecolor='none'))

    # krb
    ax.add_patch(patches.Rectangle((left_width + 0, krb_y),
                                krb_w, krb_h, 
                                linewidth=2, edgecolor='black', facecolor='none'))
    ax.text(left_width  + krb_w/2, krb_y + krb_h/2, "krb", ha='center', va='center')

    # zkosení (ručně nakreslené čáry)
    # levý spodní roh
    ax.plot([0, chamfer], [chamfer, 0], color='black', linewidth=2)
    # pravý spodní roh
    ax.plot([left_width + bottom_len - chamfer, left_width + bottom_len],
            [0, chamfer], color='black', linewidth=2)
    # levý horní roh
    ax.plot([0, 15], [left_height - 15, left_height],
            color='black', linewidth=2)

    # text popisující plochu
    #ax.text(left_width + bottom_len/2, bottom_depth + 20, "nehořlavé sklo", 
    #        color='blue', ha='center', va='center')

    # kontextová zeď
    #ax.add_patch(patches.Rectangle((left_width, krb_y + krb_h),
    #                               bottom_len, 30, 
    #                               linewidth=2, edgecolor='gray', facecolor='none'))
    #ax.text(left_width + bottom_len/2, krb_y + krb_h + 15, "zeď", ha='center', va='center')

    ax.set_aspect('equal')
    ax.set_xlim(-5, left_width + bottom_len + 5)
    ax.set_ylim(-5, left_height + 5)
    ax.axis('off')

    plt.show()
```
